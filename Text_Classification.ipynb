```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report
import nltk

nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
# Load train & test CSV files
train_df = pd.read_csv(r"c:\Users\SAM\OneDrive\Desktop\train.csv")
test_df = pd.read_csv(r"c:\Users\SAM\OneDrive\Desktop\test.csv")

print(train_df.head())
```

```python
# Combine title + description
train_df["text"] = train_df["Title"] + " " + train_df["Description"]
test_df["text"] = test_df["Title"] + " " + test_df["Description"]

# Features and labels
X_train = train_df["text"].values
y_train = train_df["Class Index"].values

X_test = test_df["text"].values
y_test = test_df["Class Index"].values

# Convert labels to 0–3 instead of 1–4
y_train = y_train - 1
y_test = y_test - 1
```

```python
vectorizer = TfidfVectorizer(stop_words=stopwords.words("english"), max_features=5000)

X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
```

```python
model = LogisticRegression(max_iter=200)
model.fit(X_train_tfidf, y_train)

# Predictions
y_pred = model.predict(X_test_tfidf)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))

print(classification_report(
    y_test, 
    y_pred, 
    target_names=["World", "Sports", "Business", "Sci/Tech"]
))
```

```python
def predict_news(text):
    vec = vectorizer.transform([text])
    pred = model.predict(vec)
    categories = ["World", "Sports", "Business", "Sci/Tech"]
    return categories[pred[0]]

sample_text = "Rohit Sharma helps India win the tournament with an incredible century and long lasting partnership."
print("Predicted Category:", predict_news(sample_text))
```

```python
import joblib

# Save model & vectorizer
joblib.dump(model, "news_classifier.pkl")
joblib.dump(vectorizer, "tfidf_vectorizer.pkl")
```
